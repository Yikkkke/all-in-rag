```python
import logging
import hashlib
from pathlib import Path
from typing import List, Dict, Any

from langchain_text_splitters import MarkdownHeaderTextSplitter
from langchain_core.documents import Document
from pathlib import Path
import uuid
```

```python
# If this code is used outside a class, replace `self.data_path` with a module-level variable.
# Use an existing `data_path` if available; otherwise fall back to the current working directory.
data_path = "../data/C8/cook/dishes"
logger = logging.getLogger(__name__)
logger.setLevel(logging.INFO)

documents = []
for filename in Path(data_path).rglob('*.md'):
    try:
        with open(filename, 'r', encoding='utf-8') as f:
            content = f.read()
            
        try:
            data_root = Path(data_path).resolve()
            relative_path = Path(filename).resolve().relative_to(data_root).as_posix()
            print(1)
            logger.info(f"Processing file: root path is {data_root}, relative path is {relative_path}")
        except Exception as e:
            relative_path = Path(filename).as_posix()
            logger.warning(f"Could not determine relative path for {filename}: {e}")
            
        # collect documents
        documents.append(Document(page_content=content, metadata={"path": str(relative_path)}))

    except Exception as e:
        logger.error(f"Error reading file {filename}: {e}")
```
